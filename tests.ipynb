```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def calculate_pt_eta_phi(df):
    df["p_l_1_pt"] = np.sqrt(df["p_l_1_x_truth"] ** 2 + df["p_l_1_y_truth"] ** 2)
    df["p_l_1_eta"] = np.arcsinh(df["p_l_1_z_truth"] / df["p_l_1_pt"])
    df["p_l_1_phi"] = np.arctan2(df["p_l_1_y_truth"], df["p_l_1_x_truth"])

    df["p_l_2_pt"] = np.sqrt(df["p_l_2_x_truth"] ** 2 + df["p_l_2_y_truth"] ** 2)
    df["p_l_2_eta"] = np.arcsinh(df["p_l_2_z_truth"] / df["p_l_2_pt"])
    df["p_l_2_phi"] = np.arctan2(df["p_l_2_y_truth"], df["p_l_2_x_truth"])

    return df

def plot_pt_eta_phi(df):
    fig, axs = plt.subplots(3, 2, figsize=(12, 15))
    sns.histplot(df["p_l_1_pt"], bins=50, ax=axs[0, 0], kde=True)
    axs[0, 0].set_title("p_l_1_pt Distribution")
    axs[0, 0].set_xlabel("p_l_1_pt")
    axs[0, 0].set_ylabel("Frequency")
    sns.histplot(df["p_l_1_eta"], bins=50, ax=axs[1, 0], kde=True)
    axs[1, 0].set_title("p_l_1_eta Distribution")
    axs[1, 0].set_xlabel("p_l_1_eta")
    axs[1, 0].set_ylabel("Frequency")
    sns.histplot(df["p_l_1_phi"], bins=50, ax=axs[2, 0], kde=True)
    axs[2, 0].set_title("p_l_1_phi Distribution")
    axs[2, 0].set_xlabel("p_l_1_phi")
    axs[2, 0].set_ylabel("Frequency")
    sns.histplot(df["p_l_2_pt"], bins=50, ax=axs[0, 1], kde=True)
    axs[0, 1].set_title("p_l_2_pt Distribution")
    axs[0, 1].set_xlabel("p_l_2_pt")
    axs[0, 1].set_ylabel("Frequency")
    sns.histplot(df["p_l_2_eta"], bins=50, ax=axs[1, 1], kde=True)
    axs[1, 1].set_title("p_l_2_eta Distribution")
    axs[1, 1].set_xlabel("p_l_2_eta")
    axs[1, 1].set_ylabel("Frequency")
    sns.histplot(df["p_l_2_phi"], bins=50, ax=axs[2, 1], kde=True)
    axs[2, 1].set_title("p_l_2_phi Distribution")
    axs[2, 1].set_xlabel("p_l_2_phi")
    axs[2, 1].set_ylabel("Frequency")
    plt.tight_layout()
```

```python
hww = pd.read_csv("data/ww_1M_MG_final_truth.csv")
length = len(hww)
hww = hww[: length // 10]
X = hww[
    [
        "p_l_1_E_truth",
        "p_l_1_x_truth",
        "p_l_1_y_truth",
        "p_l_1_z_truth",
        "p_l_2_E_truth",
        "p_l_2_x_truth",
        "p_l_2_y_truth",
        "p_l_2_z_truth",
    ]
].copy()
y = hww[
    [
        "p_v_1_E_truth",
        "p_v_1_x_truth",
        "p_v_1_y_truth",
        "p_v_1_z_truth",
        "p_v_2_E_truth",
        "p_v_2_x_truth",
        "p_v_2_y_truth",
        "p_v_2_z_truth",
    ]
].copy()

X = calculate_pt_eta_phi(X)

plot_pt_eta_phi(X)
```
